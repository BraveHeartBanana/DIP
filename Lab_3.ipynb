```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img_half = cv2.imread("/content/Lab3_half100.png", cv2.IMREAD_GRAYSCALE)
img_chess = cv2.imread("/content/Lab3_chess100.png", cv2.IMREAD_GRAYSCALE)

img_eye = cv2.imread("/content/eye.jpg", cv2.IMREAD_COLOR)
img_eye_rgb = cv2.cvtColor(img_eye, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(15,5))
plt.subplot(121)
plt.imshow(img_half, cmap = 'gray');
plt.subplot(122)
hist_half = cv2.calcHist([img_half], [0], None, [256], [0, 256])
plt.plot(hist_half); plt.xlim([0, 255]); plt.grid(); plt.show();

plt.figure(figsize=(15, 5))
plt.subplot(121)
plt.imshow(img_chess, cmap = 'gray');
plt.subplot(122)
hist_chest = cv2.calcHist([img_chess], [0], None, [256], [0, 256])
plt.plot(hist_chest); plt.xlim([0, 255]); plt.grid(); plt.show();

plt.figure(figsize=(15, 5))
plt.subplot(121)
plt.imshow(img_eye_rgb);
plt.subplot(122)
hist_eye = cv2.calcHist([img_eye_rgb], [0], None, [256], [0, 256])
plt.plot(hist_eye); plt.xlim([0, 255]); plt.grid(); plt.show();

plt.show();
```

**Застосуємо розмивання Гауса та звичайний згладжувальний фільтр до вхідного зображення**

```python
gblur_half3 = cv2.GaussianBlur(img_half,(3,3),0)
gblur_half5 = cv2.GaussianBlur(img_half,(5,5),0)
gblur_half9 = cv2.GaussianBlur(img_half,(9,9),0)
gblur_half15 = cv2.GaussianBlur(img_half,(15,15),0)

gblur_chess3 = cv2.GaussianBlur(img_chess,(3,3),0)
gblur_chess5 = cv2.GaussianBlur(img_chess,(5,5),0)
gblur_chess9 = cv2.GaussianBlur(img_chess,(9,9),0)
gblur_chess15 = cv2.GaussianBlur(img_chess,(15,15),0)


blur_img_eye_rgb5 = cv2.blur(img_eye_rgb,(5,5))
blur_img_eye_rgb15 = cv2.blur(img_eye_rgb,(15,15))
blur_img_eye_rgb30 = cv2.blur(img_eye_rgb,(30,30))
blur_img_eye_rgb60 = cv2.blur(img_eye_rgb,(60,60))
blur_img_eye_rgb100 = cv2.blur(img_eye_rgb,(100,100))
```
