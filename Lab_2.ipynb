```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img_half = cv2.imread("/content/Lab3_half100.png", cv2.IMREAD_GRAYSCALE)
img_chess = cv2.imread("/content/Lab3_chess100.png", cv2.IMREAD_GRAYSCALE)
img_chai = cv2.imread("/content/чайки.jpg", cv2.IMREAD_GRAYSCALE)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(121)
plt.imshow(img_half, cmap = 'gray');
plt.subplot(122)
hist_half = cv2.calcHist([img_half], [0], None, [256], [0, 256])
plt.plot(hist_half); plt.xlim([0, 255]); plt.grid(); plt.show();

plt.figure(figsize=(10, 5))
plt.subplot(121)
plt.imshow(img_chess, cmap = 'gray');
plt.subplot(122)
hist_chest = cv2.calcHist([img_chess], [0], None, [256], [0, 256])
plt.plot(hist_chest); plt.xlim([0, 255]); plt.grid(); plt.show();

plt.figure(figsize=(10, 5))
plt.subplot(121)
plt.imshow(img_chai, cmap = 'gray');
plt.subplot(122)
hist_half = cv2.calcHist([img_chai], [0], None, [256], [0, 256])
plt.plot(hist_half); plt.xlim([0, 255]); plt.grid(); plt.show();

plt.show();
```

## **Зміна контрасту**
 




Визначемо мінімальне та максимальне значення яскравості пікселів на вхідному зображенні. Використовуючи ці значення побудуємо графік залежності перетворення яскравостей вхідного та вихідного зображення.

```python
Lmin_half = np.min(img_half)
Lmax_half = np.max(img_half)

Lmin_chess = np.min(img_chess)
Lmax_chess = np.max(img_chess)

Lmin_chai = np.min(img_chai)
Lmax_chai = np.max(img_chai)

Loutmin = 0
Loutmax = 255

r_half = [0, Lmin_half, Lmax_half, 255]
s_half = [0, Loutmax/8, Loutmax*7/8, Loutmax]

r_chess = [0, Lmin_chess, Lmax_chess-30, 255]
s_chess = [0, Loutmax/14, Loutmax*3/14, Loutmax]

r_chai = [0, Lmin_chai, Lmax_chai-30, 255]
s_chai = [0, Loutmax/14, Loutmax*3/14, Loutmax]

plt.figure(figsize=(15, 15))
plt.subplot(221)
plt.plot(r_half, s_half, 'o-')
plt.xlim([0, 255])
plt.ylim([0, 255])
plt.ylabel('Яскравість пікселів вихідного зображення, r')
plt.xlabel('Яскравість пікселів вхідного зображення, s')
plt.grid()

plt.subplot(222)
plt.plot(r_chess, s_chess, 'o-')
plt.xlim([0, 255])
plt.ylim([0, 255])
plt.ylabel('Яскравість пікселів вихідного зображення, r')
plt.xlabel('Яскравість пікселів вхідного зображення, s')
plt.grid()

plt.subplot(223)
plt.plot(r_chai, s_chai, 'o-')
plt.xlim([0, 255])
plt.ylim([0, 255])
plt.ylabel('Яскравість пікселів вихідного зображення, r')
plt.xlabel('Яскравість пікселів вхідного зображення, s')
plt.grid()
plt.show()
```

Перше зображення

```python
Loutmax = 255
contr_img_man_half = np.zeros(img_half.shape, img_half.dtype)
alpha = int(((Loutmax*7/8) - (Loutmax/8))/ (Lmax_half - Lmin_half))
for y in range(img_half.shape[0]):
    for x in range(img_half.shape[1]):
        contr_img_man_half[y,x] = alpha*(img_half[y,x] - Lmin_half)
```

Друге зображення

```python
Loutmax = 255
contr_img_man_chess = np.zeros(img_chess.shape, img_chess.dtype)
alpha = int(((Loutmax*7/8) - (Loutmax/8))/ (Lmax_chess - Lmin_chess))
for y in range(img_chess.shape[0]):
    for x in range(img_chess.shape[1]):
        contr_img_man_chess[y,x] = alpha*(img_chess[y,x] - Lmin_chess) 
```

Третє зображення

```python
Loutmax = 255
contr_img_man_chai = np.zeros(img_chai.shape, img_chai.dtype)
alpha = int(((Loutmax*3) - (Loutmax/6))/ (Lmax_chai - Lmin_chai))
for y in range(img_chai.shape[0]):
    for x in range(img_chai.shape[1]):
        contr_img_man_chai[y,x] = alpha*(img_chai[y,x] - Lmin_chai) 
```

**Відображення**

```python
plt.figure(figsize=(10,5))
plt.subplot(121)
plt.imshow(contr_img_man_half, cmap='gray', vmin = 0, vmax = 255); 
plt.subplot(122)
hist = cv2.calcHist([contr_img_man_half], [0], None, [256], [0, 256])
plt.plot(hist); plt.xlim([0, 255]); plt.grid(); plt.show();

plt.figure(figsize=(10,5))
plt.subplot(121)
plt.imshow(contr_img_man_chess, cmap='gray', vmin = 0, vmax = 255); 
plt.subplot(122)
hist = cv2.calcHist([contr_img_man_chess], [0], None, [256], [0, 256])
plt.plot(hist); plt.xlim([0, 255]); plt.grid(); plt.show();

plt.figure(figsize=(10,5))
plt.subplot(121)
plt.imshow(contr_img_man_chai, cmap='gray', vmin = 0, vmax = 255); 
plt.subplot(122)
hist = cv2.calcHist([contr_img_man_chai], [0], None, [256], [0, 256])
plt.plot(hist); plt.xlim([0, 255]); plt.grid(); plt.show();

plt.show();
```
