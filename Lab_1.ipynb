```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_bw = cv2.imread("/voron.jpg", cv2.IMREAD_GRAYSCALE)
plt.figure(figsize=(8, 8))
plt.imshow(img_bw, cmap='gray')
plt.show()
```

**Перетворення BGR to RGB**

```python
img_eagle = cv2.imread("/eagle.jpg")

img_rgb = cv2.cvtColor(img_eagle, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(6, 6))
plt.imshow(img_eagle)
plt.show()

plt.figure(figsize=(6, 6))
plt.imshow(img_rgb)
plt.show();
```

**Перетворення RGB to HSV**

```python
img_hsv = cv2.cvtColor(img_eagle, cv2.COLOR_BGR2HSV)
plt.figure(figsize=(8, 8))
plt.imshow(img_hsv)
plt.show();
```

**Окремо відобразити колірні канали для зображень із колірними моделями RGB та HSV**

```python
plt.figure(figsize=(20, 20))

plt.subplot(131);
R = np.zeros(img_rgb.shape, dtype='uint8')
R[:, :, 0] = img_rgb[:, :, 0]
plt.imshow(R)

plt.subplot(132);
G = np.zeros(img_rgb.shape, dtype='uint8')
G[:, :, 1] = img_rgb[:, :, 1]
plt.imshow(G)

plt.subplot(133);
B = np.zeros(img_rgb.shape, dtype='uint8')
B[:, :, 2] = img_rgb[:, :, 2]
plt.imshow(B)

plt.show()
```

```python
plt.figure(figsize=(20, 20))

plt.subplot(131);
R = np.zeros(img_hsv.shape, dtype='uint8')
R[:, :, 0] = img_hsv[:, :, 0]
plt.imshow(R)

plt.subplot(132);
G = np.zeros(img_hsv.shape, dtype='uint8')
G[:, :, 1] = img_hsv[:, :, 1]
plt.imshow(G)

plt.subplot(133);
B = np.zeros(img_hsv.shape, dtype='uint8')
B[:, :, 2] = img_hsv[:, :, 2]
plt.imshow(B)

plt.show()
```

**Скопіювати область інтересу (Region of Interest) чорно-білого зображення, використовуючи наступні діапазони 0:img.shape[0]/2, 0:img.shape[1]/2.**

```python
RoI = img_bw[0: int(img_bw.shape[0]/2), 0: int(img_bw.shape[1]/2)]
plt.figure(figsize=(5, 5))
plt.imshow(RoI, cmap='gray')
plt.show()
```

**Зберегти область інтересу (Region of Interest)**

```python
cv2.imwrite("RoI.jpg", RoI)
```
